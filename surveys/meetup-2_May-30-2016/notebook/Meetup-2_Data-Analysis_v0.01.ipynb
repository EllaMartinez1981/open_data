# Meetup #2 Feedback Survey Analysis

Following the second MENA Devs meetup that occured on the **30th of May 2016**, we have distributed an online feedback survey to all attendees.

Below is a modest analysis of the feedback received given the small sample size.

### Questions asked in the survey:

```
Q1: "Overall how satisfied were you with the event?"
Q2: "... the talks?"
Q3: "... the venue?"
Q4: "... the food?"
Q5: "How did you first hear about our event?"
Q6: "How well did the event meet your expectations?"
Q7: "How likely are you to attend another one of our events?"
```

### Findings



````python
# General syntax to import specific functions in a library: 
##from (library) import (specific library function)
from pandas import DataFrame, read_csv

import numpy as np
import pandas as pd
import matplotlib as plt
import sys
import datetime as dt

# Enable inline plotting
%matplotlib inline

# Style the plots
plt.pyplot.style.use('ggplot')
````

````python
print('Python version ' + sys.version)
print('Pandas version ' + pd.__version__)
print('Matplotlib version ' + plt.__version__)
````

````python
# Loading the CSV file and transforming into a DataFrame
file_location = r'../feedback_survey.csv'
# Read the CSV file and add custom header names
raw_data = pd.read_csv(file_location, header=0, names=['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7']);
````

````python
raw_data
````

This is the raw data as answered by the survey takers. 

At a glimpse you can see that the total number of answers received is **15** and that most of the knowledge of the event came from the **Slack Group**.

````python
raw_data.describe()
````

Let's read a bit into the data and see what insights we can gather by looking at the **mean (average)** and the **standard deviation**.

By looking at the mean only, it seems that:
1. The majority is 'quite satisfied' by the overall event and they are 'highly likely' to give us another chance and attend again in the future
2. The talks were satisfactory though not great
2. There seems to be a slight issue with the talks, venue, and the food (but this is inconclusive yet before we look at the stds)

By looking at the stds, it seems that:
1. Attendees are generally in agreement that the overall event was satisfactory
2. Attendees are also in agreement regarding the talks
3. Attendees are not in agreement regarding the venue, food and expectations
4. Attendees are in agreement regarding their likelihood of attending another meetup

````python
plot = raw_data.plot(kind='area', figsize=(20,10), stacked=True, xticks=raw_data.index)
plot.set_title('Distribution')
````

````python
referral_channels = raw_data['Q5'].groupby(raw_data['Q5']).count()
rf_plot = referral_channels.plot(kind='bar', figsize=(15, 8), rot=0)
rf_plot.set_title('Referral Channels')
rf_plot.set_xlabel('How did you first hear about our event?')
````

In this instance it seems that the MENA-Devs Slack Group was more effective in communicating the event to our community.

# General Conclusions

The sample size is quite small to draw major conclusions, however, there are a few areas that we believe we can work on:

1. Identify other possible venues to host the event
2. Identify other possible options for food catering
3. Add more tech talks to the meetup agenda
4. Work more on increasing outreach through social media channels
5. Identify areas where the event did not meet the attendees expectations
